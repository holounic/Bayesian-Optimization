```python
import pandas as pd
```

```python
f_name = 'f9'
```

```python
cma_es = pd.read_csv(f'{f_name}_cma_es_doe.csv')[['precision', 'f_s']].rename(columns={'precision': 'cma_es_precision'})
cma_es_init = pd.read_csv(f'{f_name}_lhs_initial_doe.csv')[['precision', 'f_s']].rename(columns={'precision': 'lhs_init_precision'})
uniform = pd.read_csv(f'uniform_doe_{f_name}.csv')[['precision', 'f_s']].rename(columns={'precision': 'uniform_precision'})
lhs = pd.read_csv(f'lhs_doe_{f_name}.csv')[['precision', 'f_s']].rename(columns={'precision': 'lhs_precision'})
halton = pd.read_csv(f'halton_doe_{f_name}.csv')[['precision', 'f_s']].rename(columns={'precision': 'halton_precision'})
sobol = pd.read_csv(f'sobol_doe_{f_name}.csv')[['precision', 'f_s']].rename(columns={'precision': 'sobol_precision'})
stats = cma_es.merge(uniform.merge(lhs.merge(sobol.merge(halton.merge(cma_es_init)))))
stats.head()
```

# CMA-ES vs first individual (generated by LHS)

```python
cma_init = stats.loc[stats['cma_es_precision'] < stats['lhs_init_precision']]
print(f'CMA-ES improved lhs DoE on {cma_init.shape[0]}, ({100 * cma_init.shape[0] / stats.shape[0]}%) function instances')
cma_init[['f_s', 'cma_es_precision', 'lhs_init_precision']]
```

# CMA-ES generated vs Uniform

```python
cma_uniform = stats.loc[stats['cma_es_precision'] < stats['uniform_precision']]
print(f'DoE generated by CMA-ES is better than uniformly sampled DoE on {cma_uniform.shape[0]}, ({100 * cma_uniform.shape[0] / stats.shape[0]}%) function instances')
cma_uniform[['f_s', 'cma_es_precision', 'uniform_precision']]
```

# CMA-ES vs LHS

```python
cma_lhs = stats.loc[stats['cma_es_precision'] < stats['lhs_precision']]
print(f'DoE generated by CMA-ES is better than lhs sampled DoE on {cma_lhs.shape[0]} ({100 * cma_lhs.shape[0] / stats.shape[0]}%) function instances')
cma_lhs[['f_s', 'cma_es_precision', 'lhs_precision']]
```

# CMA-ES vs Sobol

```python
cma_sobol = stats.loc[stats['cma_es_precision'] < stats['sobol_precision']]
print(f'DoE generated by CMA-ES is better than sobol generated DoE on {cma_sobol.shape[0]} ({100 * cma_sobol.shape[0] / stats.shape[0]}%) function instances')
cma_sobol[['f_s', 'cma_es_precision', 'sobol_precision']]
```

# CMA-ES vs Halton

```python
cma_halton = stats.loc[stats['cma_es_precision'] < stats['halton_precision']]
print(f'DoE generated by CMA-ES is better than halton generated DoE on {cma_halton.shape[0]} ({100 * cma_halton.shape[0] / stats.shape[0]}%) function instances')
cma_halton[['f_s', 'cma_es_precision', 'halton_precision']]
```

```python
cma_is_best = stats.loc[(stats['cma_es_precision'] < stats['halton_precision']) & (stats['cma_es_precision'] < stats['uniform_precision']) & (stats['cma_es_precision'] < stats['lhs_precision']) & (stats['cma_es_precision'] < stats['sobol_precision'])]
print(f'DoE generated by CMA-ES is better than an baseline generated DoE on {cma_is_best.shape[0]} ({100 * cma_is_best.shape[0] / stats.shape[0]}%) function instances')
cma_is_best
```
